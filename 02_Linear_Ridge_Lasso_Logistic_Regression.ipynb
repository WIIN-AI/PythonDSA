```python
# Ridge Regression 
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV # Hyper parameter tunning
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
dfL= pd.DataFrame(df['data'],columns =df['feature_names'])
```

```python
YL= pd.DataFrame(df['target'],columns=['target'])
```

```python
YL['target'].value_counts() # to confirm the data is balanced or not ? if class_weight 
```

```python
# Train & Test 
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(dfL,YL, test_size = 0.33, random_state = 42)
```

```python
params = [{"C":[1,5,10]} , {"max_iter":[100,150]}]
model1 = LogisticRegression(C=100,max_iter = 100)
```

```python
model=GridSearchCV(model1,param_grid=params, scoring ="f1", cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
print(model.best_params_)
print(model.)
```

```python
y_predict_l=model.predict(X_test)
```

```python
y_predict_l
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score 
```

```python
confusion_matrix(y_predict_l,y_test)
```

```python
accuracy_score(y_predict_l,y_test)
```
